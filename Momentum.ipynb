```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
%matplotlib inline
import sys
```

```python
ABBV_csv = Path("Resources/ABBV.csv")
ABBV_data = pd.read_csv(ABBV_csv, index_col="Date",parse_dates=True, infer_datetime_format=True)


BAC_csv = Path("Resources/BAC.csv")
BAC_data = pd.read_csv(BAC_csv, index_col="Date",parse_dates=True, infer_datetime_format=True)


CVX_csv = Path("Resources/CVX.csv")
CVX_data = pd.read_csv(CVX_csv, index_col="Date",parse_dates=True, infer_datetime_format=True)
```

```python
span=10
Close = ABBV_data.copy(deep=True)

#10 day
sma = Close.rolling(window=span, min_periods=span).mean()[:10]
rest = Close[10:]
TD = pd.concat([sma, rest]).ewm(span=span, adjust=False).mean()
TD

MOMENTUM_A = 100 * (Close/TD)
#pd.set_option('display.max_rows',None)
MOMENTUM_A
```

```python
Close2 = BAC_data.copy(deep=True)
span=10
sma = Close2.rolling(window=span, min_periods=span).mean()[:10]
rest = Close2[10:]
TD2 = pd.concat([sma, rest]).ewm(span=span, adjust=False).mean()
TD2

MOMENTUM_B = 100 * (Close2/TD2)
MOMENTUM_B
```

```python
Close3 = CVX_data.copy(deep=True)

sma = Close3.rolling(window=10, min_periods=10).mean()[:10]
rest = Close3[10:]
TD3 = pd.concat([sma, rest]).ewm(span=10, adjust=False).mean()
TD3

MOMENTUM_C = 100 * (Close3/TD3)
MOMENTUM_C
```

```python
fig, (ax1,ax2) = plt.subplots(2,sharex=True)
fig.suptitle('ABBV')

ABBV_data.plot(ax=ax1)

ax1.set_ylabel('Price ($)')
ax1.get_xaxis().set_visible(False)

MOMENTUM_A.plot(ax=ax2)

ax2.set_ylabel('Momentum')
ax2.set_ylim(75,125)
ax2.axhline(100, color='b', linestyle='--')
```

```python
fig, (ax1,ax2) = plt.subplots(2, sharex=True)
fig.suptitle('BAC')

BAC_data.plot(ax=ax1)

ax1.set_ylabel('Price ($)')
ax1.get_xaxis().set_visible(False)

MOMENTUM_B.plot(ax=ax2)
ax2.set_ylim(75,125)
ax2.axhline(100, color='b', linestyle='--')
ax2.set_ylabel('Momentum')
```

```python
fig, (ax1,ax2) = plt.subplots(2, sharex=True)
fig.suptitle('CVX')

CVX_data.plot(ax=ax1)

ax1.set_ylabel('Price ($)')
ax1.get_xaxis().set_visible(False)

MOMENTUM_C.plot(ax=ax2)

ax2.set_ylabel('Momentum') 
ax2.set_ylim(75,125)
ax2.axhline(100, color='b', linestyle='--')
```
